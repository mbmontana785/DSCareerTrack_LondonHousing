#### 1.2.  Loading the data
Your data comes from the [London Datastore](https://data.london.gov.uk/): a free, open-source data-sharing portal for London-oriented datasets. 

```python
# First, make a variable called url_LondonHousePrices, and assign it the following link, enclosed in quotation-marks as a string:
# https://data.london.gov.uk/download/uk-house-price-index/70ac0766-8902-4eb5-aab5-01951aaed773/UK%20House%20price%20index.xls

url_LondonHousePrices = "https://data.london.gov.uk/download/uk-house-price-index/70ac0766-8902-4eb5-aab5-01951aaed773/UK%20House%20price%20index.xls"

# The dataset we're interested in contains the Average prices of the houses, and is actually on a particular sheet of the Excel file. 
# As a result, we need to specify the sheet name in the read_excel() method.
# Put this data into a variable called properties.  
properties = pd.read_excel(url_LondonHousePrices, sheet_name='Average price', index_col= None)
```

You might find you need to transpose your dataframe, check out what its row indexes are, and reset the index. You  also might find you need to assign the values of the first row to your column headings  . (Hint: recall the .columns feature of DataFrames, as well as the iloc[] method).

```python
#Slicing the 32 columns that represent a London borough, which are the only ones that seem necessary for this exercise.
#According to the Wikipedia link above, City of London is not a borough 
#Unnamed columns (34, 37, 47) seem to be used as separators between boroughs, regions and lastly the entire city

props_sub = properties.loc[:, "Barking & Dagenham":"Westminster"]
```

```python
props_sub.head()
```

```python
#Slicing out first row. Don't see why that's needed.
props_sub = props_sub.loc['1995-01-01':,:]
```

```python
props_sub.head()
```

```python
#Discovering that all the numbers in the dataframe are strings
props_sub.dtypes
```

```python
#Converting all numbers to floats
props_sub = props_sub.astype(float)
```

```python
props_sub.head()
```

```python
props_sub.tail()
```

```python
#Making sure all the numbers are floats
props_sub.dtypes
```

```python
#Changing floats to ints to prevent scientific notation later on
#With numbers this large, I don't see how decimals would matter.
props_sub = props_sub.astype(int)
```

```python
props_sub.head()
```

```python
props_sub.tail()
```

```python
#Slicing out rows before 2001-04-01 since the question to be answered only concerns the last 20 years
props_sub_20yrs = props_sub.loc['2001-04-01':, :]
```

```python
props_sub_20yrs.head()
```

```python
props_sub_20yrs.tail()
```

```python
#Transposing so I can do calculations on columns and add a 'difference' column
props_20yrs_trans = props_sub_20yrs.transpose()
```

```python
props_20yrs_trans.head()
```

```python
#Adding a difference column by subtracting prices of 2001-04-01 from prices of 2021-04-01
props_20yrs_trans['difference'] = props_20yrs_trans['2021-04-01'] - props_20yrs_trans['2001-04-01']
```

```python
props_20yrs_trans.head()
```

```python
#Sorting by the 'difference' column
props_20yrs_sorted = props_20yrs_trans.sort_values(by='difference', ascending=False)
```

```python
props_20yrs_sorted.head()
```

```python
#Creating Series to use for plotting
props_20yrs_diff = props_20yrs_sorted['difference']
```

```python
print(props_20yrs_diff)
```

```python
#Horizontal bar plot
plt.figure(figsize=(12, 8))
props_20yrs_diff.plot(kind='barh', title="Change in London Housing Prices since 2001")
plt.xlabel('British Pounds', fontsize=20)
plt.ylabel('Borough', fontsize=20)
plt.show()
```

```python
#Conventional bar plot
plt.figure(figsize=(12, 8))
props_20yrs_diff.plot(kind='bar', title="Change in London Housing Prices since 2001")
plt.ylabel('British Pounds', fontsize=20)
plt.xlabel('Borough', fontsize=20)
plt.show()
```
